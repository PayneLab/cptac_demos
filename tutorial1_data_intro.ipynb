## Data Availability
The goals of CPTAC as a consortium include the broad and open dissemination of cancer proteogenomic data. The timing of the a dataset's public release generally follows three stages: internal release to CPTAC investigators, public release with a publication embargo, and full public release. Each of the cancer types may be at a different data availability stage, depending on the date of data creation. In the Python <code>cptac</code> package, these three stages are dealt with as follows:

<b>Internally released data</b> requires a password to download.

<b>Embargoed release data</b> is publicly available, but prints an embargo statement every time you interact with the data.

<b>Public data</b> is fully released without restrictions.

## Downloading data

The cptac package stores the data files for each dataset on a remote server. When you first install cptac, you will have no data files. To install the latest version of the data files for a particular dataset, simply call the <code>cptac.download</code> function, passing the name of your desired dataset for the <code>dataset</code> parameter:

```python
cptac.download(dataset="endometrial")
```

## Exploring the data

Once you've downloaded a dataset, <code>cptac</code> allows you to load the dataset into a Python variable, and you can use that variable to access and work with the data. To load a particular dataset into a variable, type the name you want to give the variable, followed by <code>=</code>, and then type <code>cptac.</code> and the name of the dataset in <a href="https://en.wikipedia.org/wiki/Camel_case">UpperCamelCase</a> followed by two parentheses, e.g. <code>cptac.Endometrial()</code> or <code>cptac.Ccrcc()</code>:

```python
en = cptac.Endometrial()
```

To see what data is available, use the <code>en.list_data()</code> function. This displays the different types of data included in the dataset for this particular cancer type, each stored in a <a href="https://pandas.pydata.org/pandas-docs/stable/getting_started/dsintro.html#dataframe">pandas dataframe</a>. It also prints the dimensions of each dataframe.

```python
en.list_data()
```

# Molecular Omics

Data can be accessed through several "get" functions. For example, we can look at the proteomics data by using <code>en.get_proteomics()</code>. This returns a <a href="https://pandas.pydata.org/pandas-docs/stable/getting_started/dsintro.html#dataframe">pandas dataframe</a> containing the proteomic data. Each column in the proteomics dataframe is the quantitiative measurement for a particular protein. Each row in the proteomics dataframe is a sample of either a tumor or non-tumor from a cancer patient.

```python
proteomics = en.get_proteomics()
samples = proteomics.index
proteins = proteomics.columns
print("Samples:",samples[0:20].tolist()) #the first twenty samples
print("Proteins:",proteins[0:20].tolist()) #the first twenty proteins
```

## Dataframe values

Values in the dataframe are protein abundance values. Values that read "NaN" mean that particular sample from that patient had no data for that particular protein. For the endometrial CPTAC proteomics data, a TMT-reference channel strategy was used. A detailed description of this strategy can be found at <a href="https://www.nature.com/articles/s41596-018-0006-9">Nature Protocols</a> and also at <a href="https://www.ncbi.nlm.nih.gov/pubmed/?term=29988108">PubMed Central</a>. This strategy ratios each sample's abundance to a pooled reference. The ratio is then log transformed. Therefore positive values indicate a measurement higher than the pooled reference; negative values are lower than the pooled reference.

```python
proteomics.head()
```

As seen in <code>en.list_data()</code>, other omics data are also available (e.g. transcriptomics, copy number variation, phoshoproteomics).

The transcriptomics looks almost identical to the proteomics data, available in a pandas dataframe with the same convention. Each set of samples is consitent, meaning samples found in the endometrial proteomics data will be the same samples in all other endometrial dataframes.

```python
transcriptomics = en.get_transcriptomics()
transcriptomics.head()
```

# Clinical Data

The clinical dataframe lists clinical information for the patient associated with each sample (e.g. age, race, diabetes status, tumor size). 

```python
clinical = en.get_clinical()
clinical.head()
```

In addition to donating a tumor sample, some patients also had a normal sample taken for control and comparison. We can identify these samples by looking at samples with the value "Adjacent_normal" in the "Proteomics_Tumor_Normal" column in the clinical dataframe. For example, patient C3L-00006 provided a tumor sample, S001, and an adjacent normal sample, S105. We note that the adjacent normal samples do not have many values in the clinical columns, because they are not from a tumor and thus do not have a histological type.

```python
clinical[clinical["Patient_ID"].isin(["C3L-00006","C3L-00361","C3L-01246"])]
```

# Mutation data

Each cancer dataset contains mutation data for the cohort. The data consists of all somatic mutations found for each sample (meaning there will be many lines for each sample). Each row lists the specific gene that was mutated, the type of mutation, and the location of the mutation. This data is a direct import of a MAF file.

```python
somatic_mutations = en.get_somatic_mutation()
somatic_mutations.head()
```

# Exporting dataframes

If you wish to export a dataframe to a file, simply call the dataframe's <code>to_csv</code> method, passing the path you wish to save the file to, and the value separator you want:

```python
clinical = en.get_clinical()
clinical.to_csv(path_or_buf="clinical_dataframe.tsv", sep='\t')
```
